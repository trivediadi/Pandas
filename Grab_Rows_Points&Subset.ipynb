# Conditional Statement

```python
my_df=pd.read_csv('dog_data.csv')
my_df
```

```python
# Conditional Selection 
# > < = >= <= etc
# Suppose we want return Boolean
my_df[my_df=='WHITE']
```

```python
# Return the just the color column where the color is white
my_df[my_df == 'WHITE']['Color']
```

```python
# what if we want to return multiple columns
# Condition or
my_df[(my_df == 'BLACK') | (my_df=='WHITE')][['Color','Breed']]
```

```python
# Multiple Conditional AND
my_df[(my_df['Color']=='BROWN') & (my_df['Breed']=='AM PITT BULL MIX')][['Color','Breed']]
```

```python
# Set Index
new_ind=['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O','P','Q','R','S']
my_df['new_ind']=new_ind
my_df.set_index('new_ind',inplace=True)
my_df
```

```python
#Reset Index
my_df.head()
my_df.reset_index(inplace=True)
```

```python
my_df.drop(['new_ind','level_0','index'],axis=1,inplace=True)
```
