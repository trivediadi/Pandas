```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('dog_data.csv')
df
```

```python
#Count Distinct Values- Descending Order
df[['Color']].value_counts()
```

```python
# Count Distinct Values- Ascending Order
df[['Color']].value_counts(ascending=True)
```

```python
# OwnerZip Code Count without Nan Values
df.OwnerZip.value_counts()
```

```python
#OwnerZip Code Count with Nan Values
df['OwnerZip'].value_counts(dropna=False)
```

```python
# Get Relativ Frequency - Percentages
df.Color.value_counts(normalize=True)*100
```

```python
#Get specific value counts
df['Color'].value_counts()['WHITE']
```

```python
# Count Unique Values -size
df.groupby('Color').size()
```
