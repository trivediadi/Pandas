# Functions


```python
stuff={
    'Corporations': ['Apple','Google', 'Meta', 'Apple', 'Google', 'Meta'],
    'Employees': ['John', 'Jane', 'Jim', 'Jack', 'Jill', 'Joe'],
    'Salary': [100, 200, 300, 400, 500, 600],
}
my_df=pd.DataFrame(stuff)
```

```python
my_df
```

```python
# Create a function
def time_100(x):
    if(x<=300):
        return (x+100)*100
    else: 
        return x*100
df2=my_df.copy()
df2['Salary']=my_df['Salary'].apply(time_100)
```

```python
df2
```

```python
# Another way is Lambda function
my_df['Salary'].apply(lambda x: (x*100))
```

```python
# Applying to multiple columns
my_df[['Employees', 'Corporations']].apply(lambda x: x+'nice')
```

```python
# Sorting in columns
my_df.sort_values('Salary', ascending=False,inplace=True) 
my_df  
```
